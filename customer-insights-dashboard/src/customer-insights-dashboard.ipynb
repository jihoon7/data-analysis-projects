# Customer Insights Dashboard

## 1. 데이터생성


```python
# 데이터 생성 코드
import pandas as pd
import random
import os
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터 저장 경로
data_dir = "data/"
os.makedirs(data_dir, exist_ok=True)

# 샘플 데이터 생성
data = {
    "CustomerID": [i for i in range(1, 101)],
    "Region": [random.choice(["North", "South", "East", "West"]) for _ in range(100)],
    "PurchaseAmount": [random.randint(50, 1000) for _ in range(100)]
}

df = pd.DataFrame(data)
df.to_csv(os.path.join(data_dir, "customers.csv"), index=False)
```

## 2. 데이터 로드 및 초기 단계


```python
def load_data(file_path):
    return pd.read_csv(file_path)

# 데이터 로드
file_path = "data/customers.csv"
df = load_data(file_path)
print("데이터 로드 완료")
```

## 3. 데이터 분석 및 다음 단계


```python
def analyze_customer_data(df):
    insights = {
        "total_customers": len(df),
        "average_purchase": df["PurchaseAmount"].mean(),
        "top_region": df["Region"].value_counts().idxmax()
    }
    return insights

# 데이터 분석
insights = analyze_customer_data(df)
print("Customer Insights:", insights)
```

## 4. 시각화


```python
def visualize_segments(df):
    df["Segment"] = pd.cut(df["PurchaseAmount"],
                           bins=[0, 100, 500, 1000],
                           labels=["Low", "Medium", "High"])
    sns.countplot(x="Segment", data=df)
    plt.title("Customer Segments")
    plt.xlabel("Segment")
    plt.ylabel("Count")
    plt.show()

# 시각화
visualize_segments(df)
```
