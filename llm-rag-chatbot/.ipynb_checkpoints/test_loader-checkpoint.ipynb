```python
from modules.document_loader import load_documents
from modules.vectorizer import embed_chunks, collection
from modules.query_engine import search_similar_chunks

#기존 데이터 삭제
ids = collection.get().get('ids', [])
if ids:
    collection.delete(ids=ids)
    print("기존 데이터 삭제 완료")

# 문서 로드 및 임베딩
chunks = load_documents("data", chunk_size=200)
print("로드된 문장 수:", len(chunks))

embed_chunks(chunks)
print("새로 저장된 문서 수:", collection.count())

# DB 확인
docs_preview = collection.get(limit=5)
print("\n[저장된 예시 문장 미리보기]")
for d in docs_preview["documents"]:
    print("-", d)

# 검색 테스트
questions = [
    "연차는 몇 일이나 되나요?",
    "병가는 유급인가요?",
    "휴가는 어떻게 신청하나요?",
    "식대는 지원되나요?",
    "복지 포인트는 어디서 쓸 수 있나요?"
]

for q in questions:
    print(f"\n❓ 질문: {q}")
    results = search_similar_chunks(q, top_k=3, max_distance=0.95)
    if results:
        print("관련 문단:")
        for r in results:
            print("-", r)
    else:
        print("관련 문단을 찾지 못했어요.")
```
